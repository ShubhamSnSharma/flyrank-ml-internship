## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

---
### Answer

My lane is a **ranking** task.

The goal is to rank webpages from the highest to the lowest refresh priority so content teams know which pages to review first. This is a ranking problem because the output is an ordered list of pages rather than a prediction about whether a page belongs to a category.

```python
task_type = "Ranking"

print(f"ML Task Type: {task_type}")
print("Output: Ordered list of webpages by refresh priority")
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

---
### Answer

The model predicts the refresh priority of webpages.

Since the dataset does not contain a direct "needs refresh" label, the closest observed outcome is **trend_direction**, which records whether a page's search performance improved, remained stable, or declined over time.

Therefore, **trend_direction** is used as a proxy target. It is an **observed outcome** derived from historical search performance, not a manually defined rule or if-else condition.

```python
target = "trend_direction"

print("Proxy Target:", target)
print("Source: Observed outcome")
print("Reason: Historical search performance trend")
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

---
### Answer

Since this is a ranking task, I would evaluate the model using **Precision@50**.

This metric measures how many of the top 50 webpages recommended by the model are actually good candidates for content refresh.

A higher Precision@50 means the SEO team spends time reviewing pages that are more likely to deliver value.

Based on the baseline exercise from ML-01, a Precision@50 of around **0.70 (70%) or higher** would represent good performance because it substantially outperforms a simple rule-based baseline.

```python
success_metric = "Precision@50"
good_threshold = 0.70

print(f"Success Metric: {success_metric}")
print(f"Good Performance: Precision@50 >= {good_threshold:.2f}")
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

---
### Answer

A fixed rule cannot reliably identify which webpages should be refreshed because refresh priority depends on the interaction of many features rather than a single condition.

For example, impressions, CTR, average position, content age, trend direction, engagement, and search intent all influence whether refreshing a page will be valuable. Different combinations of these features influence refresh priority, making the relationship too complex for a fixed if-else rule.

A machine learning model can learn these complex relationships from historical data and produce more accurate rankings than a manually defined rule.

```python
reason = [
    "Many features influence refresh priority.",
    "Feature interactions are complex.",
    "No single if-else rule captures all cases.",
    "ML learns patterns from historical data."
]

print("Why ML beats a fixed rule:\n")
for r in reason:
    print("-", r)
```
