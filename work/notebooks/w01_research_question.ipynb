## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

---
**Answer**

The starter dataset contains 30,000 pages, with 16,262 (54.2%) showing declining performance and 22,006 (73.4%) meeting the measurable opportunity criteria. These numbers suggest there are far too many pages to review manually, making a ranked prioritization system useful for helping content teams focus on the pages that are most likely to benefit from review.


```python
import pandas as pd

# Load the starter dataset
df = pd.read_csv("content_refresh_anonymized.csv")

# Calculate key statistics
total_pages = len(df)
declining_pages = (df["trend_direction"] == "down").sum()
opportunity_pages = (
    (df["impressions_90d"] >= 100) &
    (df["sessions_90d"] > 0)
).sum()

print(f"Total pages: {total_pages:,}")
print(f"Declining pages: {declining_pages:,} ({declining_pages/total_pages:.1%})")
print(f"Pages with measurable opportunity: {opportunity_pages:,} ({opportunity_pages/total_pages:.1%})")
```
